```python
top_kannada_movies=pd.read_csv('Top kannada movies.csv',index_col=0)
```

```python
top_kannada_movies.head(10)
```

# sort data

```python
top_kannada_movies.sort_values(by='IMDb Rating',ascending=False)
```

```python
top_kannada_movies.to_csv('kannada_movies_list',sep=' ',header=True)
```

# selecting data


```python
top_kannada_movies_list=top_kannada_movies[['Title','IMDb Rating']]
```

```python
top_kannada_movies_list.sort_values(by='IMDb Rating',ascending=False)
```

```python
top_kannada_movies_list['IMDb Rating'].mean()
```

```python
top_kannada_movies_list['IMDb Rating'].max()
```

```python
top_kannada_movies_list['IMDb Rating'].min()
```

```python
top_kannada_movies_list['Title'].unique()
```
